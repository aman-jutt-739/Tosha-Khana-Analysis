```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Load the dataset (make sure the file path is correct)
df = pd.read_csv("Refined_TK_dataver2.csv")

# Drop missing values from 'Assessed Value'
assessed_values = df['Assessed Value'].dropna()

# Plot histogram and KDE
sns.histplot(assessed_values, kde=True, stat="density", bins=50)
plt.title("Distribution of Assessed Value")
plt.xlabel("Assessed Value")
plt.ylabel("Density")
plt.show()

# Fit normal distribution and plot it
mu, std = stats.norm.fit(assessed_values)
x = np.linspace(min(assessed_values), max(assessed_values), 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2)
plt.title(f"Fit results: mu = {mu:.2f}, std = {std:.2f}")
plt.xlabel("Assessed Value")
plt.ylabel("Density")
plt.show()
```

```python
retention_costs = df['Retention Cost'].dropna()
sns.histplot(retention_costs, kde=True, stat="density", bins=50)
plt.title("Distribution of Retention Cost")
plt.xlabel("Retention Cost")
plt.ylabel("Density")
plt.show()

mu_rc, std_rc = stats.norm.fit(retention_costs)
x = np.linspace(min(retention_costs), max(retention_costs), 100)
p = stats.norm.pdf(x, mu_rc, std_rc)
plt.plot(x, p, 'g', linewidth=2)
plt.title(f"Fit results: mu = {mu_rc:.2f}, std = {std_rc:.2f}")
plt.show()
```

```python
retained_prob = df['Retained'].value_counts(normalize=True).get('Yes', 0)
print(f"Probability of a gift being retained: {retained_prob:.2f}")
```

```python
affiliation_group = df.groupby('Affiliation')['Retained'].value_counts(normalize=True).unstack()
print("Conditional Probability of Retention by Affiliation:")
print(affiliation_group['Yes'].dropna().sort_values(ascending=False))
```

```python
category_group = df.groupby('Item Category')['Retained'].value_counts(normalize=True).unstack()
posterior = category_group['Yes'].dropna()
print("Posterior probabilities (P(Retained=Yes | Item Category)):")
print(posterior.sort_values(ascending=False))
```

```python
sns.kdeplot(df['Assessed Value'].dropna(), fill=True)
plt.title("KDE of Assessed Value")
plt.xlabel("Assessed Value")
plt.show()
```
